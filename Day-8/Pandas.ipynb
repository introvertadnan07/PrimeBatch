# Series

```python
s = pd.Series([1, 2, 3, 4, 5])
print(s)
print(type(s))
```

# Index

```python
s = pd.Series([1, 2, 3, 4, 5])

# Index
print(s[0])
print(s[2])
```

# Age Values

```python
s = pd.Series([21, 22, 37, 34, 25], index = ["Adam", "Eve", "Bob", "Charlie", "Whatson"])
print(s)
print(s["Whatson"])

print(s[2])
print(s.index)
```

# Homogenous

```python
s = pd.Series([21, 22, 37, 34, 25, "abc"], index = ["Adam", "Eve", "Bob", "Charlie", "Whatson", 24])
print(s)
print(s["Whatson"])
```

# Vectorized Operations

```python
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])

print(s1 + s2)
s1[0] = 100
print(s1)
```

# Drop

```python
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])

print(s1 + s2)
s1[0] = 100
changed_s1 = s1.drop(0)

print(s1)
print(changed_s1)
```
